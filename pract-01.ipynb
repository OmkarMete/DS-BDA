# Concept Hierachy 

```python
def fun1(value):
    if (value < 20):
        return "teenager"
    elif (value >= 20 and value < 40):
        return "young"
    elif (value >= 40 and value < 60):
        return "middle aged"
    elif (value >= 60):
        return "senior citizen"
    else:
        pass
```

```python
df["Age"] = df["Age"].apply(fun1)
```
